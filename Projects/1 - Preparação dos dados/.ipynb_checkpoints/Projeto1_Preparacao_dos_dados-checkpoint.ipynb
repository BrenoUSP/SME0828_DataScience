### 1 - Leia os dados com erro novamente, faça a limpeza e remova as duas últimas colunas.

Lendo dataset "Iris with errors" no formato CSV, colocando em um dataframe

### 2 - Leia os dados com erro novamente e insira a mediana de cada atributo onde for encontrada NaN.

Lendo dataset "Iris with errors" no formato CSV, colocando em um dataframe

```python
import pandas as pd 

# CSV file
irisErrors2 = pd.read_csv('https://raw.githubusercontent.com/BrenoUSP/SME0828_DataScience/master/Projects/data/iris-with-errors.csv', header=(0))
```

Substitui valores com '?' por np.NaN

```python
import numpy as np

irisErrors2 = irisErrors2.replace('?', np.nan)
```

Convertemos para o formato Numpy e ignoramos a última coluna (classe)

```python
arrIrisErrors = np.array(irisErrors2[irisErrors2.columns[0:irisErrors2.shape[1]-1]], dtype = float) 
```

Usamos a função nanmedian que calcula a mediana ignorando os valores NaN. Calculamos a mediana de cada atributo e armazenamos em um vetor. 

```python
medians = np.nanmedian(arrIrisErrors, axis = 0)
```

Pega os valores das medianas no vetor e coloca na posição que corresponderia no array. Utilizamos a função [nanmedian](https://numpy.org/doc/stable/reference/generated/numpy.median.html).

```python
for i in np.arange(0, arrIrisErrors.shape[0]):
    for j in np.arange(0, arrIrisErrors.shape[1]):
        if(np.isnan(arrIrisErrors[i,j]) == True): # verificar se é uma entrada com Nan
           arrIrisErrors[i,j] = medians[j] # insere a mediana

arrIrisErrors
```

### 3 - Considere os dados da Iris e mostre a distribuição de probabilidades de cada uma das variáveis após a normalização e padronização.

Leitura de csv do dataset "Iris" como Data Frame

```python
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/brenoslivio/SME0828_DataScience/master/Projects/data/iris.csv', header=(0))
```

##### Normalização dos dados

Abaixo temos um código para padronizar os dados referentes ao conjunto de dados iris.

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# considera somente os atributos, ignorando a última coluna, que contem a classe
X = np.array(data[data.columns[0:data.shape[1]-1]])

# prepara a função para transformar os dados
scaler = MinMaxScaler(feature_range=(0, 1))

# Realiza a normalização e coloca em um novo vetor
X_norm = scaler.fit_transform(X)
```

Para estudar a distribuição de probabilidades de cada uma das variáveis após normalização e padronização, será utilizada a função distplot() da biblioteca Seaborn. Abaixo, tem-se os histogramas para o conjunto de dados após a normalização entre 0 e 1. Obs.: para criar-se o número de bins nos histogramas abaixo, foi utilizada a regra de sturges, a qual é muito utilizada em estatística para definir o número de classes com o intuito de criar-se um histograma de frequências. Esta [página](https://pt.wikipedia.org/wiki/Histograma#Escolha_do_n%C3%BAmero_de_barras_pela_f%C3%B3rmula_de_Sturges) da wikipédia aborda a regra aqui discutida.

```python
import warnings
import matplotlib.pyplot as plt
import seaborn as sns

warnings.filterwarnings("ignore")
plt.figure(figsize=(15, 8))
x1 = sns.distplot(X_norm[:,0], bins = 'sturges', norm_hist = False, kde = True) 
plt.xlabel('Comprimento da sépala (cm)', fontsize=18)
plt.ylabel('Função densidade aproximada', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show(True)
```

```python
plt.figure(figsize=(15, 8))
x2 = sns.distplot(X_norm[:,1], bins = 'sturges')
plt.xlabel('Largura da sépala (cm)', fontsize=18)
plt.ylabel('Função densidade aproximada', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show(True)
```

```python
plt.figure(figsize=(15, 8))
x3 = sns.distplot(X_norm[:,2], bins = 'sturges')
plt.xlabel('Comprimento da pétala (cm)', fontsize=18)
plt.ylabel('Função densidade aproximada', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show(True)
```

```python
plt.figure(figsize=(15, 8))
x4 = sns.distplot(X_norm[:,3], bins = 'sturges')
plt.xlabel('Largura da pétala (cm)', fontsize=18)
plt.ylabel('Função densidade aproximada', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show(True)
```

Abaixo, plotou-se todas as distribuções em um mesmo gráfico

```python
plt.figure(figsize=(20, 8))

x1 = sns.distplot(X_norm[:,0], bins = 'sturges',hist = False, label = "Comprimento sépala")
x2 = sns.distplot(X_norm[:,1], bins = 'sturges',hist = False, label = "Largura sépala")
x3 = sns.distplot(X_norm[:,2], bins = 'sturges',hist = False, label = "Comprimento pétala")
x4 = sns.distplot(X_norm[:,3], bins = 'sturges',hist = False, label = "Largura pétala")

plt.xlabel('Tamanho em cm', fontsize=18)
plt.ylabel('Função densidade aproximada', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show(True)
```

### 6 - Encontre os outliers nos dados das bases Iris e BostonHouse.

```python
# Acessando-se os conjuntos de dados:

data_I = pd.read_csv('https://raw.githubusercontent.com/brenoslivio/SME0828_DataScience/master/Projects/data/iris.csv', header=(0))
data_B = pd.read_csv('https://raw.githubusercontent.com/brenoslivio/SME0828_DataScience/master/Projects/data/BostonHousing.csv', header=(0))
```

Trabalhemos com o conjunto de dados Iris primeiramente. Identifiquemos como outiliers, qualquer observação que seja considerada um outlier para alguma das variáveis de interesse (lembrando-se que a cetegoria "species" não é uma variável quantitativa e assim, não é pertinente para a análise):

```python
data_I_a = data_I.drop(data.columns[[4]], axis=1) # remoção da última coluna (species)

Q1 = data_I_a.quantile(0.25) # primeiro quartil
Q3 = data_I_a.quantile(0.75) # segundo quartil
IQR = Q3 - Q1 # distância interquartil

out = ((data_I_a < (Q1 - 1.5 * IQR)) | (data_I_a > (Q3 + 1.5 * IQR))).any(axis=1) # vetor contendo os valores booleanos
# para cada uma das observações indicando se ele é um outlier (True) ou não (False)

# vetor contendo os outliers para o conjunto de dados Iris
Iris_out = data_I[list(out)]

Iris_out
```

Agora, façamos o mesmo para o conjunto BostonHouse:

```python
Q1 = data_B.quantile(0.25) # primeiro quartil
Q3 = data_B.quantile(0.75) # segundo quartil
IQR = Q3 - Q1 # distância interquartil

out = ((data_B < (Q1 - 1.5 * IQR)) | (data_B > (Q3 + 1.5 * IQR))).any(axis=1) # vetor contendo os valores booleanos
# para cada uma das observações indicando se ele é um outlier (True) ou não (False)

# vetor contendo os outliers para o conjunto de dados BostonHouse
Boston_out = data_B[list(out)]

Boston_out
```
