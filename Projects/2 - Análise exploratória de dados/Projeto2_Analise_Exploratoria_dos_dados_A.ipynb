### 1 - Gere dados a partir de uma distribuição de Poisson. Varie a taxas $\lambda$ no intervalo [1,10] e mostre o gráfico da média em função da variância.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
import math


#gerando os dados
k = np.linspace(0,100,101)
```

```python
#lambdas no intervalo [1,10]
lambda1 = 1
lambda2 = 2
lambda3 = 3
lambda4 = 4
lambda5 = 5
lambda6 = 6
lambda7 = 7
lambda8 = 8
lambda9 = 9
lambda10 = 10
```

```python
f1 = []
f2 = []
f3 = []
f4 = []
f5 = []
f6 = []
f7 = []
f8 = []
f9 = []
f10 = []


for i in range(len(k)):
    f1.append(((lambda1**k[i])*np.exp(-1.0*lambda1))/math.factorial(k[i]))
    f2.append(((lambda2**k[i])*np.exp(-1.0*lambda2))/math.factorial(k[i]))
    f3.append(((lambda3**k[i])*np.exp(-1.0*lambda3))/math.factorial(k[i]))
    f4.append(((lambda4**k[i])*np.exp(-1.0*lambda4))/math.factorial(k[i]))
    f5.append(((lambda5**k[i])*np.exp(-1.0*lambda5))/math.factorial(k[i]))
    f6.append(((lambda6**k[i])*np.exp(-1.0*lambda6))/math.factorial(k[i]))
    f7.append(((lambda7**k[i])*np.exp(-1.0*lambda7))/math.factorial(k[i]))
    f8.append(((lambda8**k[i])*np.exp(-1.0*lambda8))/math.factorial(k[i]))
    f9.append(((lambda9**k[i])*np.exp(-1.0*lambda9))/math.factorial(k[i]))
    f10.append(((lambda10**k[i])*np.exp(-1.0*lambda10))/math.factorial(k[i]))
    
```

```python
plt.plot(k,f1,label=r'$\lambda$=1')
plt.plot(k,f2,label=r'$\lambda$=2')
plt.plot(k,f3,label=r'$\lambda$=3')
plt.plot(k,f4,label=r'$\lambda$=4')
plt.plot(k,f5,label=r'$\lambda$=5')
plt.plot(k,f6,label=r'$\lambda$=6')
plt.plot(k,f7,label=r'$\lambda$=7')
plt.plot(k,f8,label=r'$\lambda$=8')
plt.plot(k,f9,label=r'$\lambda$=9')
plt.plot(k,f10,label=r'$\lambda$=10')
plt.xlabel('k')
plt.ylabel('P(X=k)')
plt.legend()
plt.tight_layout()
```

Dado que na poisson a média é o lambda e a variância também, a medida que aumentamos o valor do lambda a curva fica mais achatada.


```python
#supondo média 10 e tamanho 100
from scipy.stats import poisson
dadospoi = poisson.rvs(mu=10, size=100)
```

```python
sns.distplot(dadospoi)
plt.xlabel('k')
plt.ylabel('P(X=k)')
```
